```python
from langchain.llms import Ollama
from langchain.tools import DuckDuckGoSearchRun
from langchain.agents import initialize_agent, AgentType
from langchain.memory import ConversationBufferMemory

# Load Llama 3 locally via Ollama
llm = Ollama(model="llama3.2")

# Enable DuckDuckGo Web Search (No API keys required)
search_tool = DuckDuckGoSearchRun()

# Memory for chat history
memory = ConversationBufferMemory(memory_key="chat_history")

# Create an agent (LLM + Web Search)
agent = initialize_agent(
    tools=[search_tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    memory=memory,
    verbose=True  # Shows intermediate steps
)

# Function to fetch scheme details
def get_scheme_details(query):
    search_results = search_tool.run(query + " government scheme details site:gov.in")  # Focuses on official sources
    formatted_query = f"""
    Based on the following search results, extract details about the government scheme and format them as follows:
    - **Scheme Name**: <name>
    - **Description**: <brief description>
    - **Eligibility Criteria**: <who can apply>
    - **Required Documents**: <list of required documents>
    - **Applicable Areas**: <states/cities covered>
    - **Benefits**: <subsidies, financial aid, etc.>
    - **Official Link**: <source URL>
    
    Search Results:
    {search_results}
    """
    return agent.run(formatted_query)

# Example usage
user_query = "Pradhan Mantri Awas Yojana eligibility"
response = get_scheme_details(user_query)
print(response)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# List of URLs and their corresponding scheme names
websites = [
    {
        "name": "Aadhar Card",
        "url": "https://www.aadhaarindia.gov.in/",
        "description_tag": "div.about-aadhaar",
        "eligibility_tag": "section#eligibility",
        "documents_tag": "section#required-documents",
        "benefits_tag": "section#benefits"
    },
    {
        "name": "PAN Card",
        "url": "https://www.incometaxindia.gov.in/pages/tax-information-services.aspx",
        "description_tag": "div#pan-description",
        "eligibility_tag": "section#eligibility",
        "documents_tag": "section#documents-required",
        "benefits_tag": "section#pan-benefits"
    },
    {
        "name": "Passport",
        "url": "https://www.passportindia.gov.in/",
        "description_tag": "div.about-passport",
        "eligibility_tag": "section#eligibility",
        "documents_tag": "section#required-documents",
        "benefits_tag": "section#passport-benefits"
    },
    {
        "name": "Pension Schemes",
        "url": "https://labour.gov.in/pension-schemes",
        "description_tag": "div.pension-description",
        "eligibility_tag": "section#eligibility",
        "documents_tag": "section#documents-required",
        "benefits_tag": "section#pension-benefits"
    },
    {
        "name": "Provident Fund",
        "url": "https://www.epfindia.gov.in/",
        "description_tag": "div.epf-scheme-description",
        "eligibility_tag": "section#eligibility",
        "documents_tag": "section#documents-required",
        "benefits_tag": "section#pf-benefits"
    },
    {
        "name": "Education Scholarships",
        "url": "https://scholarships.gov.in/",
        "description_tag": "div.scholarship-description",
        "eligibility_tag": "section#eligibility",
        "documents_tag": "section#required-documents",
        "benefits_tag": "section#benefits"
    }
]

def scrape_scheme_data(scheme):
    """Scrape relevant information for a given scheme from its website."""
    try:
        response = requests.get(scheme["url"])
        
        if response.status_code != 200:
            print(f"Failed to fetch {scheme['name']} data.")
            return None
        
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract relevant fields based on tags and classes
        description = soup.select_one(scheme["description_tag"]).get_text(strip=True) if soup.select_one(scheme["description_tag"]) else "N/A"
        eligibility = soup.select_one(scheme["eligibility_tag"]).get_text(strip=True) if soup.select_one(scheme["eligibility_tag"]) else "N/A"
        required_documents = soup.select_one(scheme["documents_tag"]).get_text(strip=True) if soup.select_one(scheme["documents_tag"]) else "N/A"
        benefits = soup.select_one(scheme["benefits_tag"]).get_text(strip=True) if soup.select_one(scheme["benefits_tag"]) else "N/A"

        scheme_data = {
            'schemeName': scheme['name'],
            'description': description,
            'eligibility': eligibility,
            'requiredDocuments': required_documents,
            'benefits': benefits,
            'officialLink': scheme['url'],
            'applicableAreas': ['India']  # Example, adjust accordingly
        }
        return scheme_data
    except Exception as e:
        print(f"Error scraping {scheme['name']}: {e}")
        return None

# Scrape all schemes
all_schemes_data = []

for scheme in websites:
    data = scrape_scheme_data(scheme)
    if data:
        all_schemes_data.append(data)

# Save the data to a CSV file
if all_schemes_data:
    df = pd.DataFrame(all_schemes_data)
    df.to_csv('government_schemes.csv', index=False)
    print("Data scraped and saved to 'government_schemes.csv'")
else:
    print("No data scraped.")
```
